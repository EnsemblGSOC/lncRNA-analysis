```python
def overlap_length(A, B):
    return(min(A[1], B[1]) - max(A[0], B[0]) + 1) 
```

```python
gene_ens_trans = []
gene_ref_trans = []
        
gene_max_exonic_count = []
gene_max_exonic_length = []
```

```python
ens_feature_ids = df_feature_file['ENS lincRNA gene ID'].tolist()
ref_feature_ids = df_feature_file['Ref lncRNA gene ID'].tolist()
```

```python
from tqdm import tqdm
```

```python
%%time
def get_max_exon_count_and_length():
    for i in tqdm(range(len(df_feature_file))):
        
        max_exon_count = 0
        max_exon_length = 0
        
        if df_feature_file['No. of transcripts in lincrna'][i] == 1:
            ens_trans_list = [df_ens_all_transcripts_index.loc[ens_feature_ids[i]]['Transcript ID']]
        
        elif df_feature_file['No. of transcripts in lincrna'][i] != 1:
            ens_trans_list = df_ens_all_transcripts_index.loc[ens_feature_ids[i]]['Transcript ID'].tolist()
            
        if df_feature_file['No. of transcripts in lncrna'][i] == 1:
            ref_trans_list = [df_ref_all_transcripts_index.loc[ref_feature_ids[i]]['RefSeq Transcript ID']]
            
        elif df_feature_file['No. of transcripts in lncrna'][i] != 1:
            ref_trans_list = df_ref_all_transcripts_index.loc[ref_feature_ids[i]]['RefSeq Transcript ID'].tolist()
#         print(ens_trans_list[0])
        for j in range(len(ens_trans_list)):
#             print(True)
            overlaps = False
            
            for k in range(len(ref_trans_list)):
#                 print(True)
                exon_count = 0
                exon_length = 0
                
                for l in range(len(df_ens_exon_index.loc[df_feature_file['ENS lincRNA gene ID'][i], ens_trans_list[j]])):
#                     print(True)
                    for m in range(len(df_ref_exon_index.loc[df_feature_file['Ref lncRNA gene ID'][i], ref_trans_list[k]])):
#                         print(True)
                        if df_ens_trans_for_trans_index.loc[ens_trans_list[j]]['No. of Exons'] == 1:
                            ens_end_list = [df_ens_exon_index.loc[df_feature_file['ENS lincRNA gene ID'][i], ens_trans_list[j]]['Exon End'].tolist()]
                            ens_start_list = [df_ens_exon_index.loc[df_feature_file['ENS lincRNA gene ID'][i], ens_trans_list[j]]['Exon Start'].tolist()]
                
                        elif df_ens_trans_for_trans_index.loc[ens_trans_list[j]]['No. of Exons'] != 1:
                            ens_start_list = df_ens_exon_index.loc[df_feature_file['ENS lincRNA gene ID'][i], ens_trans_list[j]]['Exon Start'].tolist()
                            ens_end_list = df_ens_exon_index.loc[df_feature_file['ENS lincRNA gene ID'][i], ens_trans_list[j]]['Exon End'].tolist()
                        
                        ens_coords = np.c_[ens_start_list, ens_end_list]
                
                        if df_ref_trans_for_trans_index.loc[ref_trans_list[k]]['No. of Exons'] == 1:
                            ref_start_list = [df_ref_exon_index.loc[df_feature_file['Ref lncRNA gene ID'][i], ref_trans_list[k]]['Start'].tolist()]
                            ref_end_list = [df_ref_exon_index.loc[df_feature_file['Ref lncRNA gene ID'][i], ref_trans_list[k]]['End'].tolist()]
        
                        elif df_ref_trans_for_trans_index.loc[ref_trans_list[k]]['No. of Exons'] != 1:
                            ref_start_list = df_ref_exon_index.loc[df_feature_file['Ref lncRNA gene ID'][i], ref_trans_list[k]]['Start'].tolist()
                            ref_end_list = df_ref_exon_index.loc[df_feature_file['Ref lncRNA gene ID'][i], ref_trans_list[k]]['End'].tolist()
                        
                        ref_coords = np.c_[ref_start_list, ref_end_list]
        
                        overlap = overlap_length(ens_coords[l], ref_coords[m])
                        
                        if overlap > 0:
                            overlaps = True
                            exon_length += overlap

                    if overlaps:
                        exon_count +=1
        
        
                if exon_count > max_exon_count:
                    max_ref_trans = ref_trans_list[k]
                    max_ens_trans = ens_trans_list[j]
                    max_exon_count = exon_count
                if exon_length > max_exon_length:
                    max_exon_length = exon_length    
                    
        gene_ens_trans.append(max_ens_trans)  
        gene_ref_trans.append(max_ref_trans)
        
        gene_max_exonic_count.append(max_exon_count)
        gene_max_exonic_length.append(max_exon_length)
```

```python
%%time
get_max_exon_count_and_length()
```

```python
seq_align_score = []
```

```python
def get_seq_align_score():
    for i in tqdm(range(len(df_feature_file))):
        
        if ens_seq[i] == 'Exon Count is 0':
            seq_align_score.append(0)
            continue
        
        else:
            alignments = pairwise2.align.globalxx(ens_seq[i], ref_seq[i])
            a = (list(set(alignments)))
            all_seq_score = []
            for j in range(len(a)):
                all_seq_score.append(a[j][2])
            
            seq_align_score.append(max(all_seq_score))
```

```python
%%time
get_seq_align_score()
```
