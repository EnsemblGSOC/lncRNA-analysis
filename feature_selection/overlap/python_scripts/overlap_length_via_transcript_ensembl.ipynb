```python
lincrna_gene_id = []
all_gene_id = []
lincrna_transcript_id = []
all_transcript_id = []
no_exons_lincrna = []
no_exons_all_gene = []
chr_region = []
overlap_len = []
```

```python
lincrna_id = df_all_lincrna_transcript_data['Gene ID']
gene_id = df_all_transcript_data['Gene ID']
l_t_id = df_all_lincrna_transcript_data['Transcript ID']
a_t_id = df_all_transcript_data['Transcript ID']
genes_exon_no = df_all_transcript_data['No. of Exons']
l_exon_no = df_all_lincrna_transcript_data['No. of Exons']
```

```python
def overlap(A, B):
    if A[0] < B[1] and B[0] < A[1]:
        return True
    else:
        return False
```

```python
def overlap_length(A, B):
    return(min(A[1], B[1]) - max(A[0], B[0]) + 1) 
```

```python
from tqdm import tqdm
def overlap_transcript():

    for i in tqdm(range(0,24,1)):
        
        if i == 0:
            for j in range(0,len_lincrna_gene_chr[a[i]]):  
                for k in range(0, len_all_gene_chr[a[i]]):

                    if overlap(lincrna_gene_coords[j], all_gene_coords[k]):
                        
                        overlap_len.append(overlap_length(lincrna_gene_coords[j], all_gene_coords[k]))
                        lincrna_gene_id.append(lincrna_id[j])
                        lincrna_transcript_id.append(l_t_id[j])
                        no_exons_lincrna.append(l_exon_no[j])
                        no_exons_all_gene.append(genes_exon_no[k])
                        all_gene_id.append(gene_id[k])
                        all_transcript_id.append(a_t_id[k])
                        chr_region.append(a[i])
                            
                    else:
                        
                        continue

        else:
        
            start_index_lincrna = 0
            start_index_all_gene = 0
            
            for l in range(0,i):
                start_index_all_gene = start_index_all_gene + len_all_gene_chr[a[l]]
                start_index_lincrna = start_index_lincrna + len_lincrna_gene_chr[a[l]]
            print('start lincrna {0}'.format(start_index_lincrna))
            print('end lincrna {0}'.format(start_index_lincrna + len_lincrna_gene_chr[a[i]]))
            print('start all_gene {0}'.format(start_index_all_gene))
            print('end all_gene {0}'.format(start_index_all_gene + len_all_gene_chr[a[i]]))
            
            
            for j in range(start_index_lincrna, start_index_lincrna + len_lincrna_gene_chr[a[i]], 1):
                for k in range(start_index_all_gene, start_index_all_gene + len_all_gene_chr[a[i]], 1):
                    
                    if overlap(lincrna_gene_coords[j], all_gene_coords[k]):   
                        
                        overlap_len.append(overlap_length(lincrna_gene_coords[j], all_gene_coords[k]))
                        lincrna_gene_id.append(lincrna_id[j])
                        lincrna_transcript_id.append(l_t_id[j])
                        no_exons_lincrna.append(l_exon_no[j])
                        no_exons_all_gene.append(genes_exon_no[k])
                        all_gene_id.append(gene_id[k])
                        all_transcript_id.append(a_t_id[k])
                        chr_region.append(a[i])
                            
                    else:
                        
                        continue
```

```python
%%time
overlap_transcript()
```

```python
import csv 
header = ['lincRNA gene ID', 'all gene ID', 'lincRNA trans ID', 'All trans ID', 'No. of exons all gene', 'No. of exons lincRNA', 'Chr region', 'overlap length']

path = '/all_trans_overlap.csv'

with open(path, 'wt', newline ='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(i for i in header)
```

```python
import pandas as pd

df2 = pd.read_csv(path)

df2[df2.columns[0]] = lincrna_gene_id
df2[df2.columns[1]] = all_gene_id
df2[df2.columns[2]] = lincrna_transcript_id
df2[df2.columns[3]] =all_transcript_id
df2[df2.columns[4]] = no_exons_all_gene
df2[df2.columns[5]] = no_exons_lincrna
df2[df2.columns[6]] = chr_region
df2[df2.columns[7]] = overlap_len

df2.to_csv(path)
```
