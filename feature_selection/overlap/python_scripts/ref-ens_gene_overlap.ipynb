```python
import pandas as pd
df_ens_linc_gene = pd.read_csv('/dataset/Ensembl_numeric/all_lincRNA_data/all_lincRNA_sorted_chr/all_lincrna_gene_data_sorted_chr.csv', index_col=0)
```

```python
len_ens_linc_gene_chr = df_ens_linc_gene['Seq region Name'].value_counts()
```

```python
len_ref_lnc_gene_chr = df_ref_lnc_gene['Seq_id'].value_counts()
```

```python
ens_linc_gene_strand = df_ens_linc_gene['Strand']
ref_lnc_gene_strand = df_ref_lnc_gene['Strand']
```

```python
ens_linc_gene_coords = df_ens_linc_gene[['Start','End']].to_numpy()
ref_lnc_gene_coords = df_ref_lnc_gene[['Start','End']].to_numpy()
```

```python
a = []
for i in range(0,22,1):
    a.append(str(i+1))
```

```python
a.append('X')
a.append('Y')
a[0]
```

```python
linc_gene_id = []
ref_gene_id = []
same_opp_overlap = []
no_transcripts_linc = []
no_transcripts_ref = []
chr_region = []
ens_ref_overlap = []
```

```python
for i in range(len(df_ref_lnc_gene)):
    ens_ref_overlap.append('Yes')
```

```python
lincrna_id = df_ens_linc_gene['Gene ID']
ref_id = df_ref_lnc_gene['RefSeq Gene ID']
linc_trans_no = df_ens_linc_gene['No. of Transcripts']
ref_trans_no = df_ref_lnc_gene['No. of Transcripts']
```

```python
def overlap(A, B):
    if A[0] < B[1] and B[0] < A[1]:
        return True
    else:
        return False
```

```python
from tqdm import tqdm
def overlap_gene():

    for i in tqdm(range(0,24,1)):
        
        if i == 0:
            for j in range(0,len_ens_linc_gene_chr[a[i]]):  
                for k in range(0, len_ref_lnc_gene_chr[a[i]]):

                    if overlap(ens_linc_gene_coords[j], ref_lnc_gene_coords[k]):
                           
                        no_transcripts_linc.append(linc_trans_no[j])
                        no_transcripts_ref.append(ref_trans_no[k])
                        linc_gene_id.append(lincrna_id[j])
                        ref_gene_id.append(ref_id[k])
                        chr_region.append(a[i])
                        
                        if ens_linc_gene_strand[j] == ref_lnc_gene_strand[k]:
                            same_opp_overlap.append(1)
                            
                        elif ens_linc_gene_strand[j] != ref_lnc_gene_strand[k]:
                            same_opp_overlap.append(0)
                            
                    else:
                        continue

        else:
        
            start_index_lincrna = 0
            start_index_ref = 0
            
            for l in range(0,i):
                start_index_ref = start_index_ref + len_ref_lnc_gene_chr[a[l]]
                start_index_lincrna = start_index_lincrna + len_ens_linc_gene_chr[a[l]]
                
            print('start lincrna {0}'.format(start_index_lincrna))
            print('end lincrna {0}'.format(start_index_lincrna + len_ens_linc_gene_chr[a[i]]))
            print('start ref {0}'.format(start_index_ref))
            print('end ref {0}'.format(start_index_ref + len_ref_lnc_gene_chr[a[i]]))
            
            
            for j in range(start_index_lincrna, start_index_lincrna + len_ens_linc_gene_chr[a[i]], 1):
                for k in range(start_index_ref, start_index_ref + len_ref_lnc_gene_chr[a[i]], 1):
                        
                    if overlap(ens_linc_gene_coords[j], ref_lnc_gene_coords[k]):   
                        
                        no_transcripts_linc.append(linc_trans_no[j])
                        no_transcripts_ref.append(ref_trans_no[k])
                        linc_gene_id.append(lincrna_id[j])
                        ref_gene_id.append(ref_id[k])
                        chr_region.append(a[i])
                        
                        if ens_linc_gene_strand[j] == ref_lnc_gene_strand[k]:
                            same_opp_overlap.append(1)
                            
                        elif ens_linc_gene_strand[j] != ref_lnc_gene_strand[k]:
                            same_opp_overlap.append(0)
                            
                    else:
                        continue
```

```python
%%time
overlap_gene()
```

```python
correct_ens_ref_overlap = []
for i in range(len(linc_gene_id)):
    correct_ens_ref_overlap.append(ens_ref_overlap[i])
```

```python
import csv 
header = ['ENS lincRNA gene ID', 'Ref lncRNA gene ID', 'Chr region', 'No. of transcripts in lincrna', 'No. of transcripts in lncrna', 'ENS-Ref Overlap ?', 'same_opp_overlap']

path = '/feature_selection/overlap/Ens-Ref/ens_ref_gene_overlap.csv'

with open(path, 'wt', newline ='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(i for i in header)
```
