```python
def overlap(A, B):
    if A[0] < B[1] and B[0] < A[1]:
        return True
    else:
        return False
```

```python
ens_coords = df_ens_all_exons[['Exon Start','Exon End']].to_numpy()
ref_coords = df_ref_all_exons[['Start','End']].to_numpy()
```

```python
no_exons_which_overlap = []
```

```python
for i in range(len(df_overlap)):
    no_exons_which_overlap.append('-')
```

```python
def get_no_of_overlapping_exons():
    for i in tqdm(range(len(df_overlap))):
        
        if df_overlap['same_opp_overlap'][i] == 0:
            continue
            
        else:
#         ens_exon_start = df_ens_all_exons_index.loc[ens_overlap_ids[i]]['Exon Start']
#         ens_exon_end = df_ens_all_exons_index.loc[ens_overlap_ids[i]]['Exon End']
            ens_coords_index = df_ens_all_exons.index[df_ens_all_exons['Gene ID'] == ens_overlap_ids[i]].tolist()
            ref_coords_index = df_ref_all_exons.index[df_ref_all_exons['RefSeq Gene ID'] == ref_overlap_ids[i]].tolist()
            overlap_count = 0

            for j in range(len(ens_coords_index)):
                for k in range(len(ref_coords_index)):

                    if overlap(ens_coords[ens_coords_index[j]], ref_coords[ref_coords_index[k]]):
                        overlap_count = overlap_count + 1
                        break

                    else:
                        continue

            no_exons_which_overlap[i] = overlap_count
#         for j in df_ens_all_exons_index.loc[ens_overlap_ids[i]]['Exon Start'].iteritems():
#             for k in df_ref_all_exons_index.loc[ref_overlap_ids[i]].iteritems():
                
#                 if overlap(ens_coords[j], ref_coords[k]):
#                     overlap_count = overlap_count + 1
#                     break
```

```python
%%time
get_no_of_overlapping_exons()
```
