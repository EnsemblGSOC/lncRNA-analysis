```python
import pandas as pd
df_ens_linc_genes = pd.read_csv('/dataset/Ensembl_numeric/all_lincRNA_data/all_lincRNA_sorted_chr/all_lincrna_gene_data_sorted_chr.csv', index_col=0)
```

```python
df_ens_linc_genes.head(3)
```

```python
df_ens_linc_index = df_ens_linc_genes.set_index(['Gene ID'])
```

```python
df_feature_file = pd.read_csv('/feature_selection/ORF/Ens-Ref/ens_ref_ORF&overlap_cleaned(ens&ref).csv', index_col=0)
```

```python
len(df_feature_file)
```

```python
df_feature_file.head(3)
```

```python
ens_linc_gene_ids = df_feature_file['ENS lincRNA gene ID'].tolist()
```

```python
df_ens_linc_index.loc['ENSG00000236648']['Display Name'][0:4]
```

```python
from tqdm import tqdm
```

```python
ens_linc_hgnc = []
```

```python
%%time
def is_ens_linc_hgnc():
    for i in tqdm(range(len(ens_linc_gene_ids))):
        if df_ens_linc_index.loc[ens_linc_gene_ids[i]]['Display Name'][0:4] == 'LINC':
            ens_linc_hgnc.append(1)
        else:
            ens_linc_hgnc.append(0)
```

```python
%%time
is_ens_linc_hgnc()
```

## Ref lnc HGNC count done below

```python
df_ref_lnc_genes = pd.read_csv('/dataset/refseq_numeric/all_type=lncRNA_data/all_lncrna_sorted_chr/all_lncrna_gene_data_sorted_chr.csv', index_col=0)
```

```python
df_ref_lnc_genes.head(2)
```

```python
df_ref_lnc_index = df_ref_lnc_genes.set_index(['RefSeq Gene ID'])
```

```python
ref_lnc_gene_ids = df_feature_file['Ref lncRNA gene ID'].tolist()
```

```python
df_ref_lnc_index.loc['gene-MIR1302-2HG']['Name'][0:4]
```

```python
ref_lnc_hgnc = []
```

```python
%%time
def is_ref_lnc_hgnc():
    for i in tqdm(range(len(ref_lnc_gene_ids))):
        if df_ref_lnc_index.loc[ref_lnc_gene_ids[i]]['Name'][0:4] == 'LINC':
            ref_lnc_hgnc.append(1)
        else:
            ref_lnc_hgnc.append(0)
```

```python
is_ref_lnc_hgnc()
```

```python
len(ref_lnc_hgnc)
```

```python
len(ens_linc_hgnc)
```

```python
df_feature_file['HGNC ID refseq lncrna?'] = ref_lnc_hgnc
df_feature_file['HGNC ID ensembl lincrna?'] = ens_linc_hgnc
```
