## Below function [get_refseq_gene_ids()] to get refseq gene data from gff3 file

```python
from tqdm import tqdm
```

```python
%%time
def get_refseq_gene_ids():
    for i in tqdm(range(len(annotation.df['attributes']))):
        if annotation.df['attributes'][i][0:7] == 'ID=gene':
            gene_index.append(i)
            gene_start.append(annotation.df['start'][i])
            gene_end.append(annotation.df['end'][i])
            gene_seqid.append(annotation.df['seq_id'][i])
            gene_strand.append(annotation.df['strand'][i])
            gene_type.append(annotation.df['type'][i])
            
############################_DATA_COLLECTED_FROM_ATTRIBUTES_GIVEN_BELOW_#####################################################################            
            
            gene_biotype.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('gene_biotype=')+len('gene_biotype='):annotation.df['attributes'][i].find('', len(annotation.df['attributes'][i]) if annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('gene_biotype=')+len('gene_biotype=')) == -1 else annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('gene_biotype=')+len('gene_biotype=')))])
            gene_refseq_ids.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('ID=')+len('ID='):annotation.df['attributes'][i].find('', len(annotation.df['attributes'][i]) if annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('ID=')+len('ID=')) == -1 else annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('ID=')+len('ID=')))])
            gene_dbxref.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('Dbxref=')+len('Dbxref='):annotation.df['attributes'][i].find('', len(annotation.df['attributes'][i]) if annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('Dbxref=')+len('Dbxref=')) == -1 else annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('Dbxref=')+len('Dbxref=')))])
            gene_name.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('Name=')+len('Name='):annotation.df['attributes'][i].find('', len(annotation.df['attributes'][i]) if annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('Name=')+len('Name=')) == -1 else annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('Name=')+len('Name=')))])
            gene_gbkey.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('gbkey=')+len('gbkey='):annotation.df['attributes'][i].find('', len(annotation.df['attributes'][i]) if annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('gbkey=')+len('gbkey=')) == -1 else annotation.df['attributes'][i].find(';', annotation.df['attributes'][i].find('gbkey=')+len('gbkey=')))])
            
            
#             gene_refseq_ids.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('ID=')+3:annotation.df['attributes'][i].find(";",annotation.df['attributes'][i].find('ID=')+3)]) 
#             gene_dbxref.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('Dbxref=')+7:annotation.df['attributes'][i].find(";",annotation.df['attributes'][i].find('Dbxref=')+7)])     
#             gene_name.append(annotation.df['attributes'][i][annotation.df['attributes'][i].find('Name=')+5:annotation.df['attributes'][i].find(";",annotation.df['attributes'][i].find('Name=')+5)])  
```
